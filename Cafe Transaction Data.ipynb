```python
import numpy as np
from collections import defaultdict
item_price_map = {
    'Cake': 3.0,
    'Cookie': 1.0,
    'Coffee': 2.0,
    'Juice': 3.0,
    'Salad': 5.0,
    'Sandwich': 4.0,
    'Smoothie': 4.0,
    'Tea': 1.5
}
price_to_items = defaultdict(list)
for item, price in item_price_map.items():
    price_to_items[price].append(item)


def fill_item_price(row):
    item = row['ITEM']
    price = row['PRICE_PER_UNIT']

    # Step 1 — item is missing/bad but price exists → reverse lookup
    if (pd.isna(item) or item in ['', 'Unknown', 'Error']) and pd.notna(price):
        if price in price_to_items:
            return np.random.choice(price_to_items[price]), price
        else:
            return item, price

    # Step 2 — price is missing/bad but item is valid → forward lookup
    elif (pd.isna(price) or price in ['Unknown', 'Error', '']) and item in item_price_map:
        correct_price = item_price_map[item]
        return item, correct_price

    # Step 3 — both fine, return as-is
    else:
        return item, price

cafe[['ITEM', 'PRICE_PER_UNIT']] = cafe.apply(lambda row: pd.Series(fill_item_price(row)), axis=1)
```

```python
mask_item_bad = cafe['ITEM'].isna() | cafe['ITEM'].isin(['', 'Unknown', 'Error'])

# Step 1 — recover price_per_unit from total_spent / quantity for the 51 rows
mask_derive_price = mask_item_bad & cafe['TOTAL_SPENT'].notna() & cafe['QUANTITY'].notna() & cafe['PRICE_PER_UNIT'].isna()
cafe.loc[mask_derive_price, 'PRICE_PER_UNIT'] = (
    cafe.loc[mask_derive_price, 'TOTAL_SPENT'] / cafe.loc[mask_derive_price, 'QUANTITY']
).round(2)

# Step 2 — reverse lookup item using the now-recovered price
def recover_item(row):
    item = row['ITEM']
    price = row['PRICE_PER_UNIT']
    if (pd.isna(item) or item in ['', 'Unknown', 'Error']) and pd.notna(price):
        if price in price_to_items:
            return np.random.choice(price_to_items[price])
    return item

cafe['ITEM'] = cafe.apply(recover_item, axis=1)

# Step 3 — the remaining 3 where everything is missing → random fill
mask_still_bad = cafe['ITEM'].isna() | cafe['ITEM'].isin(['', 'Unknown', 'Error'])
print("Rows still needing random fill:", mask_still_bad.sum())
```

```python
# after deriving price
cafe.loc[mask_derive_price, 'PRICE_PER_UNIT'] = (
    cafe.loc[mask_derive_price, 'TOTAL_SPENT'] / cafe.loc[mask_derive_price, 'QUANTITY']
).round(2)

# after deriving quantity
mask_derive_qty = cafe['TOTAL_SPENT'].notna() & cafe['PRICE_PER_UNIT'].notna() & cafe['QUANTITY'].isna()
cafe.loc[mask_derive_qty, 'QUANTITY'] = (
    cafe.loc[mask_derive_qty, 'TOTAL_SPENT'] / cafe.loc[mask_derive_qty, 'PRICE_PER_UNIT']
).round().astype(int)

# enforce integer type on the whole column at the end
cafe['QUANTITY'] = cafe['QUANTITY'].round().astype('Int64')
```

```python
# inspect the 6 problematic rows
mask = cafe['ITEM'].isna() | cafe['ITEM'].isin(['', 'Unknown', 'Error'])
print(cafe[mask][['ITEM', 'QUANTITY', 'PRICE_PER_UNIT', 'TOTAL_SPENT']])
```

```python
# Step 1 — random fill item + price for all 6
mask_bad = cafe['ITEM'].isna() | cafe['ITEM'].isin(['', 'Unknown', 'Error'])

def fill_unrecoverable(row):
    item = row['ITEM']
    if pd.isna(item) or item in ['', 'Unknown', 'Error']:
        random_item = np.random.choice(list(item_price_map.keys()))
        random_price = item_price_map[random_item]
        return random_item, random_price
    return item, row['PRICE_PER_UNIT']

cafe.loc[mask_bad, ['ITEM', 'PRICE_PER_UNIT']] = cafe[mask_bad].apply(
    lambda row: pd.Series(fill_unrecoverable(row)), axis=1).values

# Step 2 — recover quantity for rows 3779, 7597, 9819 using total_spent / price
mask_derive_qty = cafe['QUANTITY'].isna() & cafe['TOTAL_SPENT'].notna() & cafe['PRICE_PER_UNIT'].notna()
cafe.loc[mask_derive_qty, 'QUANTITY'] = (
    cafe.loc[mask_derive_qty, 'TOTAL_SPENT'] / cafe.loc[mask_derive_qty, 'PRICE_PER_UNIT']
).round().astype(int)

# Step 3 — verify
print(cafe[['ITEM', 'QUANTITY', 'PRICE_PER_UNIT']].isna().sum())
```

```python
# check the distribution first
print(cafe['QUANTITY'].value_counts().sort_index())
```

```python
# fix the rogue 6
cafe.loc[cafe['QUANTITY'] == 6, 'QUANTITY'] = 5

# fill the 20 missing with mode
cafe['QUANTITY'] = cafe['QUANTITY'].fillna(5).astype('Int64')

# derive TOTAL_SPENT
cafe['TOTAL_SPENT'] = cafe['QUANTITY'] * cafe['PRICE_PER_UNIT']

# verify
print("QUANTITY missing:", cafe['QUANTITY'].isna().sum())
print("TOTAL_SPENT missing:", cafe['TOTAL_SPENT'].isna().sum())
```

```python
for col in ['LOCATION', 'PAYMENT_METHOD']:
    print(cafe[col].value_counts(dropna=False))
    print()
```

```python
# fill location proportionally
loc_valid = cafe[cafe['LOCATION'].isin(['Takeaway', 'In-Store'])]['LOCATION']
loc_counts = loc_valid.value_counts(normalize=True)
loc_mask = cafe['LOCATION'].isna() | cafe['LOCATION'].isin(['Error', 'Unknown'])

cafe.loc[loc_mask, 'LOCATION'] = np.random.choice(
    loc_counts.index.tolist(),
    size=loc_mask.sum(),
    replace=True,
    p=loc_counts.values
)

# fill payment method proportionally
pay_valid = cafe[cafe['PAYMENT_METHOD'].isin(['Digital Wallet', 'Credit Card', 'Cash'])]['PAYMENT_METHOD']
pay_counts = pay_valid.value_counts(normalize=True)
pay_mask = cafe['PAYMENT_METHOD'].isna() | cafe['PAYMENT_METHOD'].isin(['Error'])

cafe.loc[pay_mask, 'PAYMENT_METHOD'] = np.random.choice(
    pay_counts.index.tolist(),
    size=pay_mask.sum(),
    replace=True,
    p=pay_counts.values
)

# verify
print(cafe['LOCATION'].value_counts())
print()
print(cafe['PAYMENT_METHOD'].value_counts())
```
